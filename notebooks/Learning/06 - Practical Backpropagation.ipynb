```python
import numpy as np

data = np.loadtxt('/aiuoa/datasets/students_1.txt', delimiter=',')
print(data)
```

```python
import keras.utils

inputs, outputs = np.hsplit(data, np.array([2]))
print(outputs)
outputs = keras.utils.to_categorical(outputs)
print(outputs)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential([
    Dense(32, input_shape=(2,)),
    Activation('relu'),
    Dense(10),
    Activation('softmax'),
    Dense(2),
    Activation('softmax'),
])
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
print(model)
```

```python
model.fit(inputs, outputs, epochs=1000, batch_size=32)
```
